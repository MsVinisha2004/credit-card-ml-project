```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
 # loading the dataset to a pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

Under Sampling

Build a sample dataset containing similar distribution of normal transaction and Fraudulent Transaction

Number of Fraudulent Transactions--> 492

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Treinig data & Testig Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation


Accuracy Score

```python
# accuracy on trainig data
X_train_prediciton = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediciton, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediciton = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediciton, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Real-time prediction function
def real_time_prediction(model):
    print(" Enter transaction details for real-time fraud detection:")

    input_data = []
    for i in range(1, 29):  # V1 to V28
        value = float(input(f"Enter V{i}: "))
        input_data.append(value)

    amount = float(input("Enter Amount: "))
    input_data.append(amount)

    # Convert input to numpy array
    input_array = np.array(input_data).reshape(1, -1)

    # Make prediction
    prediction = model.predict(input_array)

    if prediction[0] == 0:
        print(" This transaction is Legitimate.")
    else:
        print(" Fraudulent Transaction Detected!")

# Call the function
real_time_prediction(model)
```
